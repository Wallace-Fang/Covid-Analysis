```python
#import libraries
import requests
import pandas as pd
import time
```

```python
def video_statistics(video_id):
  url_video_stats = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_KEY
  stats_response = requests.get(url_video_stats).json()

  #acquire video's view, like, and comment count
  view_count = stats_response['items'][0]['statistics']['viewCount']
  like_count = stats_response['items'][0]['statistics']['likeCount']
  comment_count = stats_response['items'][0]['statistics']['commentCount']

  return view_count, like_count, comment_count
```

```python
def get_videos(df):
  API_KEY = "AIzaSyC_8f_eHAS4ieE_1okmunZcVCefKbNdF7M"

  #Channel ID for youtube channel: LinusTechTips
  CHANNEL_ID = "UCXuqSBlHAE6Xw-yeJA0Tunw"

  #order results by date, maximum results of 10000, and in case multiple pages pageToken will access those pages
  pageToken= ""
  url = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=10000"+pageToken
  response = requests.get(url).json()
  time.sleep(1)

  for video in response['items']:

    #filters out none-video data and youtube short videos
    if video['id']['kind'] == "youtube#video" and video['snippet']['title'][-7:] != "#SHORTS":

      video_id = video['id']['videoId']

      #acquire video title
      video_title = video['snippet']['title']

      #acquire video update date
      upload_date = video['snippet']['publishTime']

      #Removes the timestamp and keeps only upload date
      upload_date = upload_date.split('T')[0] 
      
      view_count, like_count, comment_count = video_statistics(video_id)

      df = df.append({"video_id": video_id,
                  "video_title":video_title, 
                  "upload_date":upload_date, 
                  "view_count":view_count, 
                  "like_count":like_count, 
                  "comment_count":comment_count}, ignore_index=True)
  return df
```

```python
df = pd.DataFrame(columns=["video_id", "video_title", "upload_date", "view_count", "like_count", "comment_count"])
df = get_videos(df)
```
